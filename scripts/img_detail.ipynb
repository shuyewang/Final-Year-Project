```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# img_full = cv2.imread("/root/autodl-tmp/DL/SCALE/DIV2K_valid_HR/cubic/0882_x2.bmp")
img_full = cv2.imread("/root/autodl-tmp/DL/DATASET/DIV2K_valid_HR/0882.png")
img_full = cv2.cvtColor(img_full, cv2.COLOR_BGR2RGB)

# plt.figure(frameon=False)
# plt.imshow(img_full)
# plt.xticks([]), plt.yticks([])
# plt.axis('off')
# plt.show()
```

```python
crop_x = 930
crop_y = 400
img_sub = img_full[crop_y:crop_y+256, crop_x:crop_x+256]

dst_size = (800, 800)

img_tmp = cv2.copyMakeBorder(img_sub, 1, 1, 1, 1, cv2.BORDER_CONSTANT)
img_tmp = cv2.resize(img_tmp, dsize=(dst_size[0],dst_size[1]), interpolation=cv2.INTER_CUBIC)

img_result = img_full

y = img_full.shape[0]-dst_size[0]
x = img_full.shape[1]-dst_size[1]
y_ = img_full.shape[0]
x_ = img_full.shape[1]

img_result[y:y_,x:x_,:] = img_tmp

plt.figure(frameon=False)
plt.imshow(img_full)
plt.xticks([]), plt.yticks([])
plt.axis('off')
plt.show()
```
